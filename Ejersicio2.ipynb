```python
# Clase que guarda los datos de un turno de espera
class TurnoEspera:
    def __init__(self, numero_turno, nombre_paciente, hora_llegada, motivo_consulta):
        self.numero_turno = numero_turno
        self.nombre_paciente = nombre_paciente
        self.hora_llegada = hora_llegada
        self.motivo_consulta = motivo_consulta


# Nodo doble, tiene anterior y siguiente
class NodoDoble:
    def __init__(self, turno):
        self.turno = turno
        self.sig = None
        self.ant = None


# Lista doble enlazada para la sala de espera
class SalaEspera:
    def __init__(self):
        self.inicio = None
        self.fin = None

    def llegada_paciente(self, turno):
        nuevo = NodoDoble(turno)

        # Si no hay pacientes, este será el primero y el último
        if self.inicio is None:
            self.inicio = nuevo
            self.fin = nuevo
        else:
            self.fin.sig = nuevo
            nuevo.ant = self.fin
            self.fin = nuevo

    def mostrar_sala(self):
        if self.inicio is None:
            print("La sala de espera está vacía.")
            return

        actual = self.inicio

        print("SALA DE ESPERA")
        print("-" * 30)

        while actual is not None:
            t = actual.turno
            print("Turno:", t.numero_turno)
            print("Paciente:", t.nombre_paciente)
            print("Hora:", t.hora_llegada)
            print("Motivo:", t.motivo_consulta)
            print("-" * 30)

            actual = actual.sig

    def buscar_turno(self, numero_turno):
        actual = self.inicio

        while actual is not None:
            if actual.turno.numero_turno == numero_turno:
                return actual

            actual = actual.sig

        return None

    def ver_siguiente(self, numero_turno):
        nodo = self.buscar_turno(numero_turno)

        if nodo is None:
            print("No se encontró ese turno.")
        elif nodo.sig is None:
            print("Este paciente no tiene siguiente.")
        else:
            print("El siguiente paciente es:", nodo.sig.turno.nombre_paciente)

    def ver_anterior(self, numero_turno):
        nodo = self.buscar_turno(numero_turno)

        if nodo is None:
            print("No se encontró ese turno.")
        elif nodo.ant is None:
            print("Este paciente no tiene anterior.")
        else:
            print("El paciente anterior es:", nodo.ant.turno.nombre_paciente)

    def prioridad_emergencia(self, numero_turno):
        nodo = self.buscar_turno(numero_turno)

        if nodo is None:
            print("No se encontró ese turno.")
            return

        if nodo == self.inicio:
            print("El paciente ya está al inicio de la lista.")
            return

        # Se desconecta el nodo de su posición actual
        if nodo.ant is not None:
            nodo.ant.sig = nodo.sig

        if nodo.sig is not None:
            nodo.sig.ant = nodo.ant
        else:
            self.fin = nodo.ant

        # Se mueve el nodo al inicio
        nodo.ant = None
        nodo.sig = self.inicio
        self.inicio.ant = nodo
        self.inicio = nodo

        print("El turno", numero_turno, "fue movido al inicio por emergencia.")


# Pruebas del ejercicio 2
sala = SalaEspera()

sala.llegada_paciente(TurnoEspera(101, "María Gómez", "08:00", "Dolor de cabeza"))
sala.llegada_paciente(TurnoEspera(102, "Pedro Ruiz", "08:15", "Fiebre"))
sala.llegada_paciente(TurnoEspera(103, "Lucía Vera", "08:30", "Dolor abdominal"))

sala.mostrar_sala()

sala.ver_siguiente(101)
sala.ver_anterior(103)

turno_emergencia = int(input("Ingrese el número de turno que será prioridad: "))
sala.prioridad_emergencia(turno_emergencia)

sala.mostrar_sala()
```
