```python
# Clase que guarda los datos del paciente monitoreado
class MonitoreoPaciente:
    def __init__(self, id_cama, nombre_paciente, frecuencia_cardiaca, estado_alerta):
        self.id_cama = id_cama
        self.nombre_paciente = nombre_paciente
        self.frecuencia_cardiaca = frecuencia_cardiaca
        self.estado_alerta = estado_alerta


# Nodo doble circular
class NodoCircular:
    def __init__(self, paciente):
        self.paciente = paciente
        self.sig = None
        self.ant = None


# Lista doble circular para monitoreo en UCI
class MonitoreoUCI:
    def __init__(self):
        self.inicio = None

    def conectar_paciente(self, paciente):
        nuevo = NodoCircular(paciente)

        # Si es el primer paciente, se apunta a sí mismo
        if self.inicio is None:
            self.inicio = nuevo
            nuevo.sig = nuevo
            nuevo.ant = nuevo
        else:
            ultimo = self.inicio.ant

            ultimo.sig = nuevo
            nuevo.ant = ultimo
            nuevo.sig = self.inicio
            self.inicio.ant = nuevo

    def mostrar_pacientes(self):
        if self.inicio is None:
            print("No hay pacientes conectados al monitoreo.")
            return

        actual = self.inicio

        print("PACIENTES EN MONITOREO UCI")
        print("-" * 30)

        while True:
            p = actual.paciente
            print("Cama:", p.id_cama)
            print("Paciente:", p.nombre_paciente)
            print("Frecuencia cardíaca:", p.frecuencia_cardiaca)
            print("Estado:", p.estado_alerta)
            print("-" * 30)

            actual = actual.sig

            if actual == self.inicio:
                break

    def desconectar_paciente(self, id_cama):
        if self.inicio is None:
            print("No hay pacientes para desconectar.")
            return

        actual = self.inicio

        while True:
            if actual.paciente.id_cama == id_cama:

                # Caso cuando solo hay un paciente
                if actual.sig == actual:
                    self.inicio = None
                else:
                    actual.ant.sig = actual.sig
                    actual.sig.ant = actual.ant

                    # Si se elimina el inicio, cambiamos el inicio
                    if actual == self.inicio:
                        self.inicio = actual.sig

                print("Paciente de la cama", id_cama, "desconectado correctamente.")
                return

            actual = actual.sig

            if actual == self.inicio:
                break

        print("No se encontró un paciente con esa cama.")

    def revisar_alertas(self):
        if self.inicio is None:
            print("No hay pacientes conectados.")
            return

        actual = self.inicio
        hay_alertas = False

        print("REVISIÓN DE ALERTAS")
        print("-" * 30)

        while True:
            p = actual.paciente

            # Función personalizada: detectar pacientes que necesitan atención
            if p.frecuencia_cardiaca > 120 or p.estado_alerta.lower() == "crítico":
                print("ALERTA EN CAMA:", p.id_cama)
                print("Paciente:", p.nombre_paciente)
                print("Frecuencia cardíaca:", p.frecuencia_cardiaca)
                print("Estado:", p.estado_alerta)
                print("-" * 30)
                hay_alertas = True

            actual = actual.sig

            if actual == self.inicio:
                break

        if hay_alertas == False:
            print("No hay pacientes en estado de alerta.")


# Pruebas del ejercicio 3
uci = MonitoreoUCI()

uci.conectar_paciente(MonitoreoPaciente(1, "Juan Molina", 80, "Estable"))
uci.conectar_paciente(MonitoreoPaciente(2, "Sofía Castro", 130, "En Observación"))
uci.conectar_paciente(MonitoreoPaciente(3, "Diego Ramírez", 95, "Crítico"))

uci.mostrar_pacientes()

uci.revisar_alertas()

cama = int(input("Ingrese el ID de cama que desea desconectar: "))
uci.desconectar_paciente(cama)

uci.mostrar_pacientes()
```
