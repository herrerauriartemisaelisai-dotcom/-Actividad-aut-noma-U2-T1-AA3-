```python
# Clase que guarda los datos de una atención médica
class AtencionMedica:
    def __init__(self, id_registro, nombre_paciente, fecha, monto_consulta):
        self.id_registro = id_registro
        self.nombre_paciente = nombre_paciente
        self.fecha = fecha
        self.monto_consulta = monto_consulta


# Nodo de la lista simple
class Nodo:
    def __init__(self, atencion):
        self.atencion = atencion
        self.sig = None


# Lista enlazada simple para el historial de atenciones
class HistorialAtenciones:
    def __init__(self):
        self.inicio = None

    def insertar_registro(self, atencion):
        nuevo = Nodo(atencion)

        # Si la lista está vacía, el nuevo nodo será el primero
        if self.inicio is None:
            self.inicio = nuevo
        else:
            actual = self.inicio

            # Avanzamos hasta llegar al último nodo
            while actual.sig is not None:
                actual = actual.sig

            actual.sig = nuevo

    def mostrar_historial(self):
        if self.inicio is None:
            print("No hay atenciones registradas.")
            return

        actual = self.inicio

        print("HISTORIAL DE ATENCIONES")
        print("-" * 30)

        while actual is not None:
            a = actual.atencion
            print("ID:", a.id_registro)
            print("Paciente:", a.nombre_paciente)
            print("Fecha:", a.fecha)
            print("Monto:", a.monto_consulta)
            print("-" * 30)

            actual = actual.sig

    def filtrar_por_monto(self, valor_minimo):
        if self.inicio is None:
            print("No hay atenciones registradas.")
            return

        actual = self.inicio
        encontrado = False

        print("CONSULTAS MAYORES A", valor_minimo)
        print("-" * 30)

        while actual is not None:
            a = actual.atencion

            if a.monto_consulta > valor_minimo:
                print("ID:", a.id_registro)
                print("Paciente:", a.nombre_paciente)
                print("Fecha:", a.fecha)
                print("Monto:", a.monto_consulta)
                print("-" * 30)
                encontrado = True

            actual = actual.sig

        if encontrado == False:
            print("No se encontraron consultas mayores a ese valor.")


# Pruebas del ejercicio 1
historial = HistorialAtenciones()

historial.insertar_registro(AtencionMedica(1, "Carlos Pérez", "10/05/2026", 25))
historial.insertar_registro(AtencionMedica(2, "Ana López", "11/05/2026", 40))
historial.insertar_registro(AtencionMedica(3, "Luis Torres", "12/05/2026", 60))

historial.mostrar_historial()

valor = float(input("Ingrese un monto mínimo para filtrar: "))
historial.filtrar_por_monto(valor)
```
